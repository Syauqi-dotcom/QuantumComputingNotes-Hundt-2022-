### Unitary Operators
1. have dimension $2^n$
2. do not change the modulus of the state vector, (norm preserving) $\braket {Uv | Uw} = (v^\dagger U^\dagger ) (Uw) = v^\dagger (U^\dagger U)w$
3. a state vector represents probabilities as probability amplitudes
4. $UU^\dagger$ = I, unitary matrices are necessarily invertible

### Base Class


```python
import numpy as np

class Operator(tensor.Tensor) : 
    """Operator are respresent by square, unitary matrices"""

    def __repr__(self) -> str:
        s = 'Operator('
        s += super().__str__().replace('\n','\n'+''*len(s))
        s += '/'
        return s
    
    def __str__(self) -> str :
        s = f'Operator for {self.nbits}-qubit state space.'
        s += 'Tensor: \n'
        s += super().__str__()
        return s

    def adjoint(self) -> Operator:
        return Operator(np.conj(self.transpose()))
```
